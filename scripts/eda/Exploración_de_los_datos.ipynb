## **2.5. Relación Entre Variables**
---


```python
df_heart_2022_no_nans.info()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df_numeric = df_heart_2022_no_nans.select_dtypes(include=['float64'])

# Matriz de correlación de Pearson
corr_matrix = df_numeric.corr()

# Visualizar la matriz de correlación con un heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de correlación de Pearson')
plt.show()
```

```python
# Calcular la matriz de correlación de Spearman
corr_matrix_spearman = df_numeric.corr(method='spearman')

# Visualizar la matriz de correlación de Spearman con un heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(corr_matrix_spearman, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de correlación de Spearman')
plt.show()
```

La correlación tanto por el método de Pearson como en el de Spearman, muestran mucha semejanza. Se encuentra una asociación lineal positiva débil entre MentalHealthDays y PhysicalHealthDays.

De igual manera, sse encuentra una asociación lineal moderada entre el peso en kilogramos y la altura en metros.

Por último, se muestra una asociación lineal positiva fuerte entre el peso en Kilogramos y el BMI (Body Mass Index). Esto puede deberse a que el BMI se calcula haciendo uso del peso en kilogramos, teniendo cierta redundancia en la información.

```python
# Calcular los cuantiles
q1 = df_heart_2022_clean["BMI"].quantile(0.25)
q3 = df_heart_2022_clean["BMI"].quantile(0.75)

# Calcular el rango intercuartílico (IQR)
iqr = q3 - q1

# Calcular los límites superior e inferior
upper_limit = q3 + 1.5 * iqr
lower_limit = q1 - 1.5 * iqr

print("Límite superior:", upper_limit)
print("Límite inferior:", lower_limit) # Pasa a ser 0.
# Se filtran los datos atípicos
outliers = df_heart_2022_clean["BMI"][(df_heart_2022_clean["BMI"] < lower_limit) | (df_heart_2022_clean["BMI"]> upper_limit)]

# Obtengamos la proporción de atípicos

prop_atipicos = len(outliers)/df_heart_2022_clean.shape[0]*100
prop_atipicos
```
